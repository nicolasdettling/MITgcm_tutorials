```python
import numpy as np
import matplotlib.pyplot as plt
import os
from xmitgcm import open_mdsdataset
import cmocean as co
import seaborn as sns

sns.set_context(context='paper', font_scale=1.5)
```

```python
bdir = '/Users/ndettling/mitgcm/temp/MITgcm/verification/tutorial_plume_on_slope_Hoggs/run/'
experiment0 = 'run_000'
experiment1 = 'run_001'
experiment2 = 'run_002'
experiment3 = 'run_003'
```

### Import experiment results

```python
import warnings
warnings.filterwarnings('ignore')
# suppress numerous warnings only when reading data into the workspace with xmitgcm.open_mdsdataset
with warnings.catch_warnings():
    # this will suppress all warnings in this block
    warnings.simplefilter("ignore")
    
    ds0 = open_mdsdataset(os.path.join(bdir, experiment0),
                          delta_t=20.,
                          ref_date="2001-1-1 0:0:0",
                          prefix='T',
                          geometry='cartesian')
    
    ds1 = open_mdsdataset(os.path.join(bdir, experiment1),
                          delta_t=20.,
                          ref_date="2001-1-1 0:0:0",
                          prefix='T',
                          geometry='cartesian')
    
    ds2 = open_mdsdataset(os.path.join(bdir, experiment2),
                          delta_t=20.,
                          ref_date="2001-1-1 0:0:0",
                          prefix='T',
                          geometry='cartesian')
    
    ds3 = open_mdsdataset(os.path.join(bdir, experiment3),
                          delta_t=20.,
                          ref_date="2001-1-1 0:0:0",
                          prefix='T',
                          geometry='cartesian')
```

```python
theta0 = ds0.T.where(ds0.T != 0)
theta1 = ds1.T.where(ds1.T != 0)
theta2 = ds2.T.where(ds2.T != 0)
theta3 = ds3.T.where(ds3.T != 0)
```

### Plot results at early stage

```python
time_index = 5
vmin = 9.3
vmax = 10
props = dict( facecolor='white', alpha=1)

plt.figure(num=1,figsize=[16,14])

plt.subplot(4,1,1)

ph0 = plt.pcolormesh(theta0.XC, theta0.Z, theta0.isel(time=time_index).squeeze(), cmap=co.cm.thermal, vmin=vmin, vmax=vmax)
plt.ylabel('z [m]')
cbar = plt.colorbar(ph0, label='$\\theta$ [$\degree$C]');
plt.text(150, -500, 'Fr = 0',  fontsize=18, bbox=props)

plt.subplot(4,1,2)

ph1 = plt.pcolormesh(theta1.XC, theta1.Z, theta1.isel(time=time_index).squeeze(), cmap=co.cm.thermal, vmin=vmin, vmax=vmax)
plt.ylabel('z [m]')
cbar = plt.colorbar(ph1, label='$\\theta$ [$\degree$C]');
plt.text(150, -500, 'Fr = 0.1',  fontsize=18, bbox=props);

plt.subplot(4,1,3)

ph2 = plt.pcolormesh(theta2.XC, theta2.Z, theta2.isel(time=time_index).squeeze(), cmap=co.cm.thermal, vmin=vmin, vmax=vmax)
plt.ylabel('z [m]')
cbar = plt.colorbar(ph2, label='$\\theta$ [$\degree$C]')
plt.text(150, -500, 'Fr = 0.2',  fontsize=18, bbox=props);

plt.subplot(4,1,4)

ph3 = plt.pcolormesh(theta3.XC, theta3.Z, theta3.isel(time=time_index).squeeze(), cmap=co.cm.thermal, vmin=vmin, vmax=vmax)
plt.xlabel('x [m]')
plt.ylabel('z [m]')
cbar = plt.colorbar(ph3, label='$\\theta$ [$\degree$C]')
plt.text(150, -500, 'Fr = 0.28',  fontsize=18, bbox=props);
```

#### Observations:

- gravity current propagates down-slope
- as expected, gravity wave is more intense the higher the Froude number
- for higher Froude numbers, the height of the gravity current "head" decreases
